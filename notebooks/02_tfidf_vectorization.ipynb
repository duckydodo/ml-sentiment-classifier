```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df = pd.read_csv("../data/cleaned/imdb_cleaned.csv")
df.head()
```

```python
X = df["clean_review"]
y = df["label"]
```

```python


tfidf_uni = TfidfVectorizer(
    max_features=5000,
    ngram_range=(1, 1)

)

X_tfidf_uni = tfidf_uni.fit_transform(X)
X_tfidf_uni.shape
#unigram only uses single words, does not correctly recognise possible phrases such as "not good"
```

```python
tfidf_uni.get_feature_names_out()[:20]
#ig we are using noun only lemmatization, does not affect adjectives
```

```python
from nltk.stem import WordNetLemmatizer

l = WordNetLemmatizer()

print(l.lemmatize("abandoned"))        # noun (default)
print(l.lemmatize("abandoned", pos="v"))
print(l.lemmatize("movies"))
```

```python
tfidf_bi = TfidfVectorizer(
    max_features=5000,
    ngram_range=(1, 2)
)

X_tfidf_bi = tfidf_bi.fit_transform(X)
X_tfidf_bi.shape
```

```python
tfidf_bi.get_feature_names_out()[:20]
#bi features contain both unigrams and bigrams, both one word and two word phrases
```

```python
uni_features = set(tfidf_uni.get_feature_names_out())
bi_features = set(tfidf_bi.get_feature_names_out())

len(uni_features), len(bi_features)
```

```python
list(bi_features - uni_features)
#shows everything in bi features but not uni features so only bigrams
#list is not ordered
```

```python
import pickle

with open("../data/cleaned/tfidf_unigram.pkl", "wb") as f:
    pickle.dump(tfidf_uni, f)

with open("../data/cleaned/tfidf_bigram.pkl", "wb") as f:
    pickle.dump(tfidf_bi, f)
```
