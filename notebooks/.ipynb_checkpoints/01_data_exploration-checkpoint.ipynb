```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import seaborn as sns
```

```python
df=pd.read_csv("../data/raw/IMDB Dataset.csv")
```

```python
df.shape
```

```python
df.info
```

```python
df['sentiment'].value_counts()
```

```python
df['sentiment'].value_counts().plot(
    kind='bar',
    title='Sentiment Class Distribution'
)
plt.show()
#our dataset is balanced in terms of positive and negative reviews
```

```python
df['label'] = df['sentiment'].map({'negative': 0, 'positive': 1})
df[['sentiment', 'label']].head()
df['review_length'] = df['review'].apply(lambda x: len(x.split()))
df['review_length'].describe()
```

```python
plt.hist(df['review_length'],bins=50)
plt.xlabel("Number of Words")
plt.ylabel("Frequency")
plt.title("Review Length Distribution")
plt.show()
#most reviews tend to be around 200-250 words long
```

```python
df.groupby('sentiment')['review_length'].mean()
#both sentiments have around the same average number of words
```

```python
df.boxplot(column='review_length', by='sentiment')
plt.suptitle("")
plt.title("Review Length by Sentiment")
plt.show()
#negative reviews tend to be shorter, no one writes two thousand word review for a movie that they didn't like
```
